```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

```python
eco = pd.read_csv('economics.csv')
eco.head()
```

```python
fig = plt.figure( figsize=(12,5) )

ax = plt.gca()
plt.plot(eco.date,eco.psavert, c='red')

ax.tick_params(axis='y',labelcolor = 'red')
ax.set_ylabel('Personal Savings Rate', color='red',fontsize=15)

#Create a twin y axis that share the same x axis
ax_right = ax.twinx()
ax_right.plot(eco.date, eco.unemploy)
ax_right.tick_params(axis='y',labelcolor = 'tab:blue')
ax_right.set_ylabel('# Unemployed (1000s)', color='tab:blue',fontsize=15)

ax.set_xticks(np.arange(0,len(eco.date),60))

plt.title('Personal Savings Rate vs Unemployment', fontsize=20)
plt.show()
```
